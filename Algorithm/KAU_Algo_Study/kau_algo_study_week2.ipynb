## [같은 수로 만들기(2374)](https://www.acmicpc.net/problem/2374)

### 풀이

일단 문제를 해결하기 위해서, 데이터의 전처리가 필요하다. <br>
N이 10이고, input => [10, 10, 10, 2, 2, 2, 7, 7, 7, 7]로 주어질 때, 문제 조건을 보면 굳이 중복된 수가 계속 존재할 필요가 없다.


```python
# 입력 및 데이터 전처리
n = int(input())
ls = [int(input())]
for i in range(n-1):
    tmp = int(input())

    if (ls[-1] == tmp):
        continue
    
    ls.append(tmp)
```

그렇기 때문에 우리는 위의 코드처럼 중복되는 값은 continue로 넘겨서 값을 필터링 할 것이다. <br>
ls = [10, 2, 7]이 되었다.

```python
# stack을 순회하며 조건에 맞을 시 차이만큼 cnt로 누적
stack = []
cnt = 0
for item in ls:
    while stack and stack[-1] < item:
        top = stack.pop()
        if not stack:
            cnt += item - top
        else:
            cnt += min(stack[-1], item) - top
        
    stack.append(item)


# 남아있는 stack의 잔여 데이터 cnt로 누적
while len(stack) > 1:
    top = stack.pop()
    cnt += stack[-1] - top
```

일단 기본적으로 입력받은 데이터는 sort 상태가 아니기 때문에, 필터링을 해도 해당 경향성은 존재한다. <br>
그렇기 때문에 stack을 위와 같이 활용한다면, [10, 2, 7]과 같은 데이터를 처리할 수 있게 된다. <br>
ls를 보면 2 -> 7 -> 10으로 가야 Add()을 최소로 호출할 수 있음을 알 수 있다. 즉, next는 item, top은 2, prev는 stack[-1](pop 이후)로 간주해서 보면 <br>
else 로직을 위와 같이 작성한 이유를 알 수 있을 것이다.

그리고 남아있는 stack 잔여 데이터는 내림차순인 것은 자명하기에, (stack = [10, 7], cnt = 5) <br>
위와 같이 로직을 구현하면 된다.

### 전체 코드

```python
# 해결 코드
import sys

input = sys.stdin.readline


# 입력 및 데이터 전처리
n = int(input())
ls = [int(input())]
for i in range(n-1):
    tmp = int(input())

    if (ls[-1] == tmp):
        continue
    
    ls.append(tmp)


# stack을 순회하며 조건에 맞을 시 차이만큼 cnt로 누적
stack = []
cnt = 0
for item in ls:
    while stack and stack[-1] < item:
        top = stack.pop()
        if not stack:
            cnt += item - top
        else:
            cnt += min(stack[-1], item) - top
        
    stack.append(item)


# 남아있는 stack의 잔여 데이터 cnt로 누적
while len(stack) > 1:
    top = stack.pop()
    cnt += stack[-1] - top


# 최종 출력
print(cnt)
```

## [UCPC는 무엇의 약자일까?(15904)](https://www.acmicpc.net/problem/15904)

### 풀이

풀이라고 할 것도 없다. target = "UCPC"과 입력받은 inp_str을 하나씩 대조하면서 총 4개 일치하면 break하고 I love UCPC 아니면 I hate UCPC 출력하면 된다.

### 전체 코드

```python
import sys

input = sys.stdin.readline

inp_str = input()
target = "UCPC"

idx = 0
for c in inp_str:
    if c == target[idx]:
        idx += 1
    if idx == 4:
        break

if idx == 4:
    print("I love UCPC")
else:
    print("I hate UCPC")
```

## [문자열 폭발(9935)](https://www.acmicpc.net/problem/9935)

### 풀이

이 문제는 크게 어려울 것이 없다. str_len은 최대 10^6이고, bomb_len은 최대 36이니 $O(NM)$ 로직으로 구현해도 AC 받을 수 있다. <br>
문자열의 문자를 하나씩 받아오며, stack에 저장한다. 그리고 stack의 끝부터 bomb을 체크하면서 순회하면 해결할 수 있다.

문제가 간단하니 바로 전체 코드를 보면 될 듯 하다.

### 전체 코드

```python
import sys

input = sys.stdin.readline


# 데이터 입력 및 전처리
inp_str = list(input())
inp_str.pop()   # '\n' 제거

bomb = list(input())
bomb.pop()      # '\n' 제거

bomb_len = len(bomb)


# solve
stack = []
for c in inp_str:
    stack.append(c)
    if(len(stack) >= bomb_len):
        if(stack[-bomb_len:] == bomb):
            for _ in range(bomb_len):
                stack.pop()

if stack:
    print(''.join(stack))
else:
    print("FRULA")
```
