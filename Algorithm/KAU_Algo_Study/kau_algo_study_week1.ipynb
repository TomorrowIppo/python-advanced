## [부녀회장이 될테야(2775)](https://www.acmicpc.net/problem/2775)

## 풀이

| 층 \ 호 | 1호 | 2호 | 3호 | 4호 | 5호 | ... |
| :--- | :---: | :---: | :---: | :---: | :---: | :---: |
| **0층** | 1 | 2 | 3 | 4 | 5 | ... |
| **1층** | 1 | 3 | 6 | 10 | 15 | ... |
| **2층** | 1 | 4 | 10 | 20 | 35 | ... |
| **3층** | 1 | 5 | 15 | 35 | 70 | ... |
| **...층** | ... | ... | ... | ... | ... | ... |

[input] <br>
2 <br>
1 <br>
3 <br>
2 <br>
3 <br>

설명이라고 할 것이 없다. 조건에 맞게 위의 자료처럼 미리 테이블을 만들어 둔 뒤, 입력에 맞춰 indexing하면 된다.

## 전체 코드

```python
import sys

input = sys.stdin.readline


# 1. 최대 범위(14층 14호)까지 미리 계산
apt = [[0] * 15 for _ in range(15)]


# 0층 초기화 (0층 i호에는 i명이 삼)
for i in range(1, 15):
    apt[0][i] = i


# DP 테이블 채우기
for k in range(1, 15):
    for n in range(1, 15):
        # k층 n호 = k층 n-1호(옆집) + k-1층 n호(아랫집)
        apt[k][n] = apt[k][n-1] + apt[k-1][n]


# 2. 테스트 케이스 처리
T = int(input())
for _ in range(T):
    k = int(input())
    n = int(input())
    print(apt[k][n]) # 미리 계산된 값을 O(1)에 조회
```

## [팰린드롬인지 확인하기(10988)](https://www.acmicpc.net/problem/10988)

## 풀이

풀이라고 할 것도 없다. 개행문자를 날린 뒤 원본과 뒤집은 것을 비교하면 된다.

## [단어 뒤집기 2(17413)](https://www.acmicpc.net/problem/17413)

## 풀이

이 문제는 아주 간단하다. 문자열을 입력받고, 처음부터 끝까지 순회하면서 

- '<'를 만났으면 일단 스택을 비우며 모두 ans 리스트에 저장한 뒤, '>'를 만날 때까지 모든 문자를 ans 리스트에 저장한다. 
- ' '를 만났으면 일단 스택을 비우며 모두 ans 리스트에 저장한 뒤, ' '를 ans 리스트에 저장한다.
- '<', ' ' 문자를 만나지 않는다면 stack에 저장한다.

최종으로 잔여 stack을 비우며 모두 ans 리스트에 저장한 뒤, print(''.join(ans)) 하면 정답이다. <br>
길게 설명할 부분은 없고, 직접 위의 로직을 입력 샘플에 적용하면서 생각해보면 금방 이해될 것이다.

## 전체 코드

```python
import sys
input = sys.stdin.readline

inp_str = input().strip()

ans = []
stack = []
i = 0
while i < len(inp_str):
    if inp_str[i] == '<':
        while stack:
            ans.append(stack.pop())

        while True:
            if inp_str[i] == '>':
                ans.append('>')
                break

            ans.append(inp_str[i])
            i += 1
    
    elif inp_str[i] == ' ':
        while stack:
            ans.append(stack.pop())
        
        ans.append(' ')
    
    else:
        stack.append(inp_str[i])
    
    i += 1

while stack:
    ans.append(stack.pop())

print(''.join(ans))
```

## [한국이 그리울 땐 서버에 접속하지(9996)](https://www.acmicpc.net/problem/9996)

## 풀이

사실 엄청 간단한 문제이지만, 까다로운 조건이 하나 있다.


첫째 줄에 파일의 개수 N이 주어진다. (1 ≤ N ≤ 100)

둘째 줄에는 패턴이 주어진다. 패턴은 알파벳 소문자와 별표(아스키값 42) 한 개로 이루어져 있다. 문자열의 길이는 100을 넘지 않으며, 별표는 문자열의 시작과 끝에 있지 않다.

다음 N개 줄에는 파일 이름이 주어진다. 파일 이름은 알파벳 소문자로만 이루어져 있고, 길이는 100을 넘지 않는다.

패턴의 길이가 파일 이름의 길이보다 길 수도 있다는 것을 인지하지 않고 문제를 풀면 무조건 에러가 뜰 것이다. <br>
그것을 염두하고 문제를 풀면 크게 어려울 것이 없는 문제이다.

## 전체 코드

```python
import sys

input = sys.stdin.readline

n = int(input())
pattern = input().strip()

prefix, suffix = pattern.split('*')
p_len = len(prefix)
s_len = len(suffix)

for _ in range(n):
    inp_str = input().strip()

    if len(inp_str) < (p_len + s_len):
        print("NE")
        continue

    if inp_str[:p_len] == prefix and inp_str[-s_len:] == suffix:
        print("DA")
    else:
        print("NE")
```
