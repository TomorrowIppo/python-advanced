## [부녀회장이 될테야(2775)](https://www.acmicpc.net/problem/2775)

## 풀이

| 층 \ 호 | 1호 | 2호 | 3호 | 4호 | 5호 | ... |
| :--- | :---: | :---: | :---: | :---: | :---: | :---: |
| **0층** | 1 | 2 | 3 | 4 | 5 | ... |
| **1층** | 1 | 3 | 6 | 10 | 15 | ... |
| **2층** | 1 | 4 | 10 | 20 | 35 | ... |
| **3층** | 1 | 5 | 15 | 35 | 70 | ... |
| **...층** | ... | ... | ... | ... | ... | ... |

[input] <br>
2 <br>
1 <br>
3 <br>
2 <br>
3 <br>

설명이라고 할 것이 없다. 조건에 맞게 위의 자료처럼 미리 테이블을 만들어 둔 뒤, 입력에 맞춰 indexing하면 된다.

## 전체 코드

```python
import sys

input = sys.stdin.readline


# 1. 최대 범위(14층 14호)까지 미리 계산
apt = [[0] * 15 for _ in range(15)]


# 0층 초기화 (0층 i호에는 i명이 삼)
for i in range(1, 15):
    apt[0][i] = i


# DP 테이블 채우기
for k in range(1, 15):
    for n in range(1, 15):
        # k층 n호 = k층 n-1호(옆집) + k-1층 n호(아랫집)
        apt[k][n] = apt[k][n-1] + apt[k-1][n]


# 2. 테스트 케이스 처리
T = int(input())
for _ in range(T):
    k = int(input())
    n = int(input())
    print(apt[k][n]) # 미리 계산된 값을 O(1)에 조회
```

## [팰린드롬인지 확인하기(10988)](https://www.acmicpc.net/problem/10988)

## 풀이

풀이라고 할 것도 없다. 개행문자를 날린 뒤 원본과 뒤집은 것을 비교하면 된다.

## 전체 코드

```python
import sys


# strip()을 써서 개행 문자를 날리고 문자열 그대로 사용
s = sys.stdin.readline().strip()


# 슬라이싱은 문자열에서도 바로 작동하며, 리스트 변환보다 빠릅니다.
if s == s[::-1]:
    print(1)
else:
    print(0)
```
