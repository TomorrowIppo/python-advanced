# 항공대 문제해결기법 2주차

## 1번 - [스트릿 코딩 파이터(23348)](https://www.acmicpc.net/problem/23348)

```python
import sys

input = sys.stdin.readline

level = list(map(int, input().split()))
n = int(input())
grades = []
for i in range(n):
    grade = 0
    for j in range(3):
        a = list(map(int, input().split()))
        for k in range(3):
            grade += a[k] * level[k]
    grades.append(grade)
print(max(grades))
```
