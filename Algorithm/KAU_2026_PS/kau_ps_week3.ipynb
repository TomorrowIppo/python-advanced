# 항공대 문제해결기법 3주차

## 1번 - [별 찍기 - 10(2447)](https://www.acmicpc.net/problem/2447)

## 풀이

문제만 보면 무슨 말인지 이해가 안 될 수 있다. 그럴 경우에는 N에 따라 나오는 도형을 나열해 규칙을 파악하는 것이 중요하다.<br>
우선 $N = 1$인 경우를 보자.

```python
"""
*
"""
```

다음으로 $N = 3$인 경우는 다음과 같다.

```python
"""
* * *
*   *
* * *
"""
```

다음으로 $N=9$인 경우는 다음과 같다.

```python
"""

* * * * * * * * *
*   * *   * *   *
* * * * * * * * *
* * *       * * *
*   *       *   *
* * *       * * *
* * * * * * * * *
*   * *   * *   *
* * * * * * * * *

"""
```

$N=1$, $N=3$, 그리고 $N=9$의 케이스를 다음과 같이 생각해보자

```python
"""
*

N = 1
"""

"""
* * *   => A
*   *   => B
* * *   => A

N = 3
"""

"""
* * * * * * * * *
*   * *   * *   *   => A
* * * * * * * * *

* * *       * * *
*   *       *   *   => B
* * *       * * *

* * * * * * * * *
*   * *   * *   *   => A
* * * * * * * * *

N = 9
"""
```

위의 규칙성을 보면 알 수 있듯이 <br>
재귀적으로 N // 3 하면서 결과값에 맞는 문자집합을 불러오고 그것을 A, B, A 순서에 맞게 나열하면서 리턴하도록 구현하면 된다.

## 전체 코드

```python
# boj-2447

import sys

input = sys.stdin.readline

def draw_stars(n):
    # Base Case: n이 1이면 별 하나만 담긴 리스트 반환
    if n == 1:
        return ["*"]

    # 재귀적으로 n/3 크기의 별 패턴을 가져옴
    stars = draw_stars(n // 3)
    res = []

    # 1. 상단 (3줄 중 첫 번째 줄 덩어리)
    for s in stars:
        res.append(s * 3)
    
    # 2. 중단 (3줄 중 가운데 줄 덩어리 - 가운데가 비어있음)
    for s in stars:
        res.append(s + " " * (n // 3) + s)
    
    # 3. 하단 (3줄 중 마지막 줄 덩어리 - 상단과 동일)
    for s in stars:
        res.append(s * 3)

    return res

N = int(input())
# 리스트에 담긴 문자열들을 줄바꿈으로 합쳐서 한 번에 출력
print('\n'.join(draw_stars(N)))
```

## 2번 - [칸토어 집합(4779)](https://www.acmicpc.net/problem/4779)

## 풀이

이런 문제 풀 때는 직접 그려보며 패턴을 파악하는 것이 제일 빠르다.

$N=1$ <br> --- <br> - -

$N=2$ <br> --------- <br> ---   --- <br> - -   - -

필자가 마크다운 사용을 잘 못해서 어색하게 나오는데, 어쨌든 규칙은 문자열을 저장하는 리스트가 있다고 할 때,

$ls[i] = ls[i - 1] + "공백"*pow(3, i - 1) + ls[i - 1]$

위와 같은 점화식으로 bottom-up으로 구현하면 된다. 마크다운으로 공백을 표현하려고 하는데 알아서 붙여버려서 "공백"이라고 작성했다.

## 전체 코드

```python
# boj-4779

import sys

input = sys.stdin.readline

ls = [
    "-",
    "- -",
]

for i in range(2, 13):
    ls.append(ls[i - 1] + " "*pow(3, i-1) + ls[i - 1])

while True:
    try:
        N = int(input())

        rst = ls[N]
        print(rst)
    except:
        break
```
