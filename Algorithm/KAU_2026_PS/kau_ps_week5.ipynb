## 1번 - [N-Queen(9663)](https://www.acmicpc.net/problem/9663)

## 풀이

문제는 어려워 보이지만, 생각보다 할만 하다.

### 핵심 아이디어: 백트래킹 (Backtracking)
1. **행(Row) 단위 배치:** 퀸을 한 행에 하나씩만 놓는다. (가로 공격 방지)
2. **유망함(Promising) 판단:** 현재 위치에 퀸을 놓을 수 있는지 확인한다.
   - **세로 확인:** 같은 열에 퀸이 있는가?
   - **대각선 확인:** 대각선상에 퀸이 있는가?
3. **가지치기(Pruning):** 만약 놓을 수 없다면 더 이상 진행하지 않고 되돌아간다.

### ① 위치 표현: `row = [0] * n`
- 2차원 배열 대신 **1차원 배열**을 사용하여 메모리를 아낀다.
- `row[x] = i`의 의미: **"x번째 행의 퀸은 i번째 열에 있다"**는 뜻이다.

### ② 유망성 검사: `is_promising(x)`
현재 `x`행에 놓은 퀸이 이전의 퀸들과 충돌하는지 확인한다.
- **`row[x] == row[i]`**: 같은 열(Vertical)에 있는지 확인한다.
- **`abs(row[x] - row[i]) == abs(x - i)`**: **대각선** 확인법이다. 
  - (행의 차이)와 (열의 차이)가 같다면 대각선상에 있는 것이다.

### ③ 재귀 호출: `n_queens(x)`
- `x == n`에 도달하면 모든 퀸을 성공적으로 놓은 것이므로 `ans`를 1 증가시킨다.
- 그렇지 않다면 `0`부터 `n-1`까지 열을 순회하며 퀸을 놓아보고, 가능하면 다음 행(`x+1`)으로 넘어간다.

## 전체 코드

```python
# boj-9663

import sys

input = sys.stdin.readline

n = int(input())

ans = 0
row = [0] * n

def is_promising(x):
    for i in range(x):
        if row[x] == row[i] or abs(row[x] - row[i]) == abs(x - i):
            return False
    
    return True

def n_queens(x):
    global ans
    if x == n:
        ans += 1
        return
    else:
        for i in range(n):
            row[x] = i
            if is_promising(x):
                n_queens(x + 1)

n_queens(0)
print(ans)
```

## 2번 - [LCS 2(9252)](https://www.acmicpc.net/problem/9252)

## 풀이

할만 하다.

### 핵심 아이디어: 역추적 (Backtracking)
1. 우선 9251번과 동일한 방식으로 **DP 테이블**을 완성한다.
2. 테이블의 가장 오른쪽 아래($dp[N][M]$)에서 시작하여 **역으로 거슬러 올라가며** 문자를 찾는다.
3. 이동 규칙:
   - 현재 위치의 문자가 서로 같다면? → **결과에 추가**하고 **대각선 위**로 이동.
   - 문자가 다르다면? → DP 값이 변하지 않은 방향(**위쪽** 혹은 **왼쪽**)으로 이동.

## 전체 코드

```python
# boj-9252

import sys
input = sys.stdin.readline

# 1. 입력 및 초기화 (개행 문자 제거 필수)
s1 = [""] + list(input().strip())
s2 = [""] + list(input().strip())

n1, n2 = len(s1), len(s2)
dp = [[0] * n2 for _ in range(n1)]

# 2. LCS 길이 구하기 (숫자만 저장하여 속도와 메모리 확보)
for i in range(1, n1):
    for j in range(1, n2):
        if s1[i] == s2[j]:
            dp[i][j] = dp[i-1][j-1] + 1
        else:
            dp[i][j] = max(dp[i-1][j], dp[i][j-1])

ans_len = dp[-1][-1]
print(ans_len)

# 3. 역추적 (문자열 찾아내기)
if ans_len > 0:
    result = []
    r, c = n1 - 1, n2 - 1
    
    while r > 0 and c > 0:
        # 문자가 같다면 공통 수열에 포함시키고 대각선 위로 이동
        if s1[r] == s2[c]:
            result.append(s1[r])
            r -= 1
            c -= 1
        # 위쪽 값이 현재 값과 같다면 위로 이동
        elif dp[r-1][c] == dp[r][c]:
            r -= 1
        # 왼쪽 값이 현재 값과 같다면 왼쪽으로 이동
        else:
            c -= 1
            
    # 역순으로 담겼으므로 뒤집어서 출력
    print("".join(reversed(result)))
```
