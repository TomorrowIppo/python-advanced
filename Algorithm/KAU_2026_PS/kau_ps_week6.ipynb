## 1번 - [수열의 변화(1551)](https://www.acmicpc.net/problem/1551)

## 풀이

이건 뭐 어려울 거 없다. 조건에 맞춰서 구현하면 된다.

## 전체 코드

```python
import sys

input = sys.stdin.readline

N, K = map(int, input().split())
ls = list(map(int, input().split(',')))

new_ls = []
_len = len(ls)
for _ in range(K):
    new_ls.clear()
    for i in range(_len - 1):
        new_ls.append(ls[i+1] - ls[i])
    _len = len(new_ls)
    ls = new_ls.copy()

new_str = ""
if K == 0:
    new_ls = ls

for i in range(len(new_ls)):
    new_str += str(new_ls[i])
    if(i != len(new_ls) - 1):
        new_str += ","
    
print(new_str)
```

## 2번 - [괄호 끼워넣기(11899)](https://www.acmicpc.net/problem/11899)

## 풀이

문제가 너무 쉽다. 그냥 기본적으로 stack으로 저장하되 괄호가 완성되는 경우 pop()해주면서 문자열 순회를 진행해주면서 잔여 stack을 pop() 하면서 cnt를 증가시켜주면 그것이 정답이다.

## 전체 코드

```python
import sys

input = sys.stdin.readline

inp = list(input().strip())
stack = []

for c in inp:
    if c == ')':
        if stack and stack[-1] == '(':
            stack.pop()
        else:
            stack.append(c)
    else:
        stack.append(c)
    
cnt = 0
while stack:
    stack.pop()
    cnt += 1
print(cnt)
```

## 3번 - [탑 보기(22866)](https://www.acmicpc.net/problem/22866)

## 풀이

이 문제는 어려워 보이지만, 생각보다 할만하다.

왼쪽에서부터 한 번, 오른쪽에서부터 한 번으로 총 두 번 순회하면 된다. 그리고 순회 때마다 현 위치의 번호를 스택에 저장한다.

왼쪽에서부터 진행하는 루프를 생각해보자. <br>
그럼 각 스택에 저장된 정보는 현 위치로부터 왼쪽의 건물들의 정보이다. 그럼 우린 스택이 차있으면서 현 위치의 높이보다 낮으면 pop()을 진행하면 된다. <br>
손으로 직접 스택 상태를 그려가면서 생각해보면 이해가 될 것이다.

마찬가지로 오른쪽도 똑같이 구현하되, 시야에서 보이는 건물 중 가장 가까운 건물을 선택해야 하므로 왼쪽에서 루프의 결과로 이미 데이터가 존재한다면 abs() 처리해서 짧은 거리의 위치를 채택하고 <br>
아니면 top을 넣어주면 된다.

## 전체 코드

```python
import sys

input = sys.stdin.readline

N = int(input())
heights = [0] + list(map(int, input().split()))
ans = [[0, -int(1e9)] for _ in range(N + 1)]


# 왼쪽에서부터
stack = []
for i in range(1, N + 1):
    while stack and heights[stack[-1]] <= heights[i]:
        stack.pop()

    ans[i][0] += len(stack)
    if stack:
        ans[i][1] = stack[-1]
    stack.append(i)


# 오른쪽에서부터
stack = []
for i in range(N, 0, -1):
    while stack and heights[stack[-1]] <= heights[i]:
        stack.pop()

    ans[i][0] += len(stack)
    if stack:
        if ans[i][1] != -int(1e9):
            left = abs(ans[i][1] - i)
            right = abs(i - stack[-1])

            if left > right:
                ans[i][1] = stack[-1]
        else:
            ans[i][1] = stack[-1]
    stack.append(i)

for i in range(1, N + 1):
    if ans[i][1] == -int(1e9):
        print(0)
    else:
        print(ans[i][0], ans[i][1])
```
