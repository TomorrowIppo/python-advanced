## 1번 - [부녀회장이 될테야(2775)](https://www.acmicpc.net/problem/2775)

## 풀이

| 층 \ 호 | 1호 | 2호 | 3호 | 4호 | 5호 | ... |
| :--- | :---: | :---: | :---: | :---: | :---: | :---: |
| **0층** | 1 | 2 | 3 | 4 | 5 | ... |
| **1층** | 1 | 3 | 6 | 10 | 15 | ... |
| **2층** | 1 | 4 | 10 | 20 | 35 | ... |
| **3층** | 1 | 5 | 15 | 35 | 70 | ... |
| **...층** | ... | ... | ... | ... | ... | ... |

[input] <br>
2 <br>
1 <br>
3 <br>
2 <br>
3 <br>

설명이라고 할 것이 없다. 조건에 맞게 위의 자료처럼 미리 테이블을 만들어 둔 뒤, 입력에 맞춰 indexing하면 된다.

## 전체 코드

```python
# boj-2775

import sys

input = sys.stdin.readline

T = int(input())

for i in range(T):
    k = int(input())
    n = int(input())
    people = [i for i in range(1, n+1)]

    for x in range(k):
        for y in range(1, n):
            people[y] += people[y-1]
    print(people[-1])
```

## 2번 - [가장 큰 정사각형(1915)](https://www.acmicpc.net/problem/1915)

## 풀이

사실 어려울 것이 없는 문제이다. 순회를 하며 현 위치가 1일 때, 왼쪽, 위쪽, 좌측상단 모두 1이라면 3개의 dp table 값 중에서 최솟값 + 1을 채택하며 순회하면 된다.

최솟값을 선택하는 이유는 우리가 구하는 것이 결국은 " 정사각형 " 이기 때문이다.

기본적으로 내가 설정한 $dp[i][j]$의 의미는 (i, j)를 꼭짓점으로 하는 최대 정사각형의 한 변의 길이라는 점을 생각하면 이해가 될 것이다.

## 전체 코드

```python
# boj-1915

import sys

input = sys.stdin.readline


n, m = map(int, input().split())
dp = [[0]*m for _ in range(n)]
arr = [list(map(int, input().strip())) for _ in range(n)]
ans = 0

for i in range(0, n):
    for j in range(0, m):
        if arr[i][j] == 1:
            if i == 0 or j == 0:
                dp[i][j] = 1
            else:
                if arr[i-1][j] == 1 and arr[i-1][j-1] == 1 and arr[i][j-1] == 1:
                    dp[i][j] = min(dp[i-1][j], dp[i][j-1], dp[i-1][j-1]) + 1
                else:
                    dp[i][j] = 1
        ans = max(dp[i][j], ans)

print(ans**2)
```

## 3번 - [LCS(9251)](https://www.acmicpc.net/problem/9251)

## 풀이

참 DP 문제는 직관적이지 않아서 어렵다. 하지만, LCS도 잘 생각해보면 쉬운 문제이다.

$dp[i][j]$를 $S1$의 $i$번째 문자까지와 $S2$의 $j$번째 문자까지 고려했을 때의 LCS 길이라고 정의하자.

두 문자열 $S1$과 $S2$를 비교하며 DP 테이블을 채운다.

1. **문자가 같을 때 ($S1[i-1] == S2[j-1]$):**
   - $dp[i][j] = dp[i-1][j-1] + 1$ (대각선 왼쪽 위 값 + 1)
2. **문자가 다를 때 ($S1[i] \neq S2[j]$):**
   - $dp[i][j] = \max(dp[i-1][j], dp[i][j-1])$ (위쪽과 왼쪽 중 큰 값)

| | (공백) | C | A | P | C | A | K |
| :---: | :---: | :---: | :---: | :---: | :---: | :---: | :---: |
| **(공백)** | 0 | 0 | 0 | 0 | 0 | 0 | 0 |
| **A** | 0 | 0 | **1** | 1 | 1 | 1 | 1 |
| **C** | 0 | **1** | 1 | 1 | **2** | 2 | 2 |
| **A** | 0 | 1 | **2** | 2 | 2 | **3** | 3 |
| **Y** | 0 | 1 | 2 | 2 | 2 | 3 | 3 |
| **K** | 0 | 1 | 2 | 2 | 2 | 3 | **4** |
| **P** | 0 | 1 | 2 | **3** | 3 | 3 | 4 |

**결과:** 테이블의 가장 오른쪽 아래 값인 **4**가 최장 공통 부분 수열의 길이다.

## 전체 코드

```python
# boj-9251

import sys

# input을 정의할 때 strip()을 포함하거나, 읽을 때 붙여주세요.
input = sys.stdin.readline

# 1. .strip()으로 양 끝 공백/줄바꿈 제거
s1 = input().strip()
s2 = input().strip()

n1 = len(s1)
n2 = len(s2)

# 2. DP 테이블 초기화
lcs = [[0] * (n2 + 1) for _ in range(n1 + 1)]

# 3. LCS 알고리즘 수행
for i in range(1, n1 + 1):
    for j in range(1, n2 + 1):
        if s1[i-1] == s2[j-1]:
            lcs[i][j] = lcs[i-1][j-1] + 1
        else:
            lcs[i][j] = max(lcs[i-1][j], lcs[i][j-1])

# 4. 결과 출력 (가장 마지막 칸이 최댓값임이 보장됨)
print(lcs[n1][n2])
```

## 4번 - [1로 만들기(1463)](https://www.acmicpc.net/problem/1463)

## 풀이

전체 코드의 주석을 확인할 것

## 전체 코드

```python
# boj-1463

import sys

# 1. 입력 최적화 (sys.stdin.readline 사용)
input = sys.stdin.readline

# 데이터가 하나만 들어오므로 strip() 없이도 정수 변환 가능
x = int(input())

# 2. DP 테이블 초기화 (1-based index를 위해 x+1 크기)
# d[i]는 숫자 i를 1로 만드는 최소 연산 횟수를 저장
d = [0] * (x + 1)

# 3. Bottom-Up DP 수행 (2부터 x까지)
for i in range(2, x + 1):
    # (1) 먼저 1을 빼는 경우를 기본값으로 설정 (연산 1회 추가)
    d[i] = d[i - 1] + 1
    
    # (2) 2로 나누어 떨어지면, '2로 나누기'와 '1 빼기' 중 최솟값 선택
    if i % 2 == 0:
        d[i] = min(d[i], d[i // 2] + 1)
        
    # (3) 3으로 나누어 떨어지면, 현재 값과 '3으로 나누기' 중 최솟값 선택
    # if-elif를 쓰지 않는 이유는 6처럼 2와 3으로 동시에 나눠지는 경우를 모두 체크하기 위함
    if i % 3 == 0:
        d[i] = min(d[i], d[i // 3] + 1)
        
# 4. 결과 출력
print(d[x])
```
