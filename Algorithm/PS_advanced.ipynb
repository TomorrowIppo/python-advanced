## 1. 반복문(Loop) 심화: 더 파이썬답게, 더 빠르게
단순히 for i in range(n)만 쓰는 단계를 넘어, 가독성과 성능을 모두 잡는 테크닉입니다.

### (1) List Comprehension
단순히 리스트를 만드는 것을 넘어, 조건문과 중첩 반복문을 한 줄로 처리합니다. 일반 append보다 속도가 빠릅니다.

```python
# 0부터 10까지 짝수의 제곱만 담은 리스트
evens_squared = [i**2 for i in range(11) if i % 2 == 0]

# 2차원 배열 초기화 (PS 필수!)
# n행 m열을 0으로 초기화
# matrix = [[0] * m for _ in range(n)]
```

### (2) enumerate와 zip 활용
인덱스가 필요할 때 range(len(array))를 쓰는 대신 enumerate를 쓰세요.

```python
# 인덱스와 갑승ㄹ 동시에!
for idx, value in enumerate(['A', 'B', 'C']):
    print(f"인덱스 {idx}의 값은 {value}")

# 두 개 이상의 리스트를 동시에 순회할 때
names = ['kim', 'lee', 'park']
scores = [85, 92, 78]
for name, score in zip(names, scores):
    print(f"{name}: {score}")
```

## 2. 자료구조 심화: 상황별 최적의 도구 선택
파이썬의 기본 list는 만능 같지만, 특정 연산에서는 매우 느립니다.

### (1) collections.deque (큐/스택)
리스트의 pop(0)은 $O(N)$이지만, deque의 popleft()는 $O(1)$입니다. BFS구현 시 필수입니다.

### (2) dict와 set (해시 테이블)
" 특정 원소가 있는지 확인 "하는 작업 (if x in data)을 할 때, 리스트는 $O(N)$이지만 set이나 dict는 $O(1)$입니다.
- 중복 제거가 필요하거나, 데이터의 존재 여부만 체크할 때는 무조건 set를 쓰세요.

### (3) heapq (우선순위 큐)
다익스트라(최단 경로) 알고리즘이나 "실시간으로 최솟값/최댓값을 계속 찾아야 하는 문제"에서 사용합니다.

```python
import heapq
pq = []
heapq.heappush(pq, 10)  # O(lg N)
heapq.heappop(pq)       # O(lg N)
```
