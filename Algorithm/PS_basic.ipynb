## 2. 필수 내장 함수 & 자료구조
파이썬의 강점은 '바퀴를 다시 발명할 필요가 없다'는 것입니다.

### 정렬 (Sorting)
- data.sort(): 원본 리스트 정렬
- sorted(data): 정렬된 새로운 리스트 반환
- **Key 사용**: data.sort(key=lambda x: (x[0], -x[1])) (첫 번째 인자 오름차순, 두 번째 인자 내림차순)

### 자료구조 (Collections)
- deque: 큐(Queue)와 덱(Deque) 구현 시 필수. pop(0)은 $O(N)$이지만 popleft()는 $O(1)$입니다.
- Counter: 리스트 내 요소의 개수를 세어 딕셔너리 형태로 반환합니다.

```python
from collections import deque, Counter

# Deque 사용법
q = deque([1, 2, 3])
q.append(4)
q.popleft()

# Counter 사용법
cnt = Counter([1, 1, 2, 2, 2, 3]) # {2: 3, 1: 2, 3: 1}
```

### 우선순위 큐 (Heap)
- 최소 힙(Min-heap)이 기본입니다. 최대 힙은 값에 -를 붙여 저장하세요.

```python
import heapq
heap = []
heapq.heappush(heap, 5)
val = heapq.heappop(heap)
```

```python
from itertools import permutations, combinations

data = ['A', 'B', 'C']
res1 = list(permutations(data, 2)) # 순열
res2 = list(combinations(data, 2)) # 조합
```

## 5.주의할 점 (Pitfalls)
- **복사 주의**: 리스트 복사 시 b = a는 주소 복사입니다. 실제 값 복사는 b = a[:]나, copy.deepcopy(a)를 사용하세요. (Deepcopy는 매우 느리니 신중히 사용)
- **문자열 합치기**: + 연산자로 문자열을 계속 더하는 것은 느립니다. ''.join(list)를 사용하세요.
